```python
gender_df.reset_index(drop=True, inplace=True)
```

```python
myList = []
for i in range(len( gender_df.description)):
    
    # Metindeki harf olmayan karakterler boşluk karakteriyle değiştirilir
    sentence = re.sub("[^a-zA-Z]"," ", gender_df.description[i])
    sentence = sentence.lower()
    
     # Metin boşluk karakterlerine göre parçalanır
    sentence = sentence.split()
    
    # Her bir kelimenin kökü bulunur ve stopwords listesinde olmayan kelimeler alınır
    sentence = [porter.stem(j) for j in sentence if not j in set(stopwords.words("english"))]
    
    # Kökü bulunan kelimeler bir araya getirilerek tek bir metin dizesi oluşturulur,
    sentenceson =" ".join(sentence)
    myList.append(sentenceson)
    
myList

```

```python
from sklearn.feature_extraction.text import CountVectorizer


cv = CountVectorizer(max_features=11194)
# Metin verileri CountVectorizer'a uygulanarak vektörlere dönüştürülür ve numpy dizisine dönüştürülür
x = cv.fit_transform(myList).toarray()

# Hedef değişken (y), veri çerçevesinden iloc yöntemiyle seçilir ve numpy dizisine dönüştürülür
y = gender_df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=41)
```

```python
from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
gn.fit(X_train, y_train)
y_preds = gn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_preds)
print(cm)
```

```python
print(gn.score(X_test, y_test))
```

```python
def preprocess_text(text):
    text = re.sub(r"[^\w\s]", "", text)  
    text = text.lower()  
    return text
```

```python
def predict_gender(text):
    processed_text = preprocess_text(text)
    
    vectorized_text = cv.transform([processed_text]).toarray()
    
    predicted_class = gn.predict(vectorized_text)
    
    return predicted_class[0]
```
